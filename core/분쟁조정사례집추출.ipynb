```python
import fitz
from pathlib import Path
from PIL import Image
```

```python
pdf_path = Path("./dataset/2021_주택임대차분쟁조정사례집.pdf")
```

```python
doc = fitz.open(pdf_path)
img_dir = Path("pages")
img_dir.mkdir(exist_ok=True)
```

```python
for i, page in enumerate(doc):
    pix = page.get_pixmap(dpi=300)  # 300dpi 중요
    img_path = img_dir / f"page_{i+1:03}.png"
    pix.save(img_path)
```

```python
import pytesseract
from pathlib import Path
from PIL import Image
```

```python
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"
```

```python
texts=[]
for img_path in sorted(img_dir.glob("*.png")):
    img = Image.open(img_path)
    text = pytesseract.image_to_string(
        img,
        lang="kor",
        config="--psm 6"
    )
    texts.append(f"\n\n===== {img_path.name} =====\n{text}")
    
out = Path("cases_ocr.txt")
out.write_text("".join(texts), encoding="utf-8")
print("OCR 완료:", out)
```
