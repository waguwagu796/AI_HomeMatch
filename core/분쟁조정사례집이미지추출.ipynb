```python
import fitz  # PyMuPDF
from pathlib import Path

# PDF 파일 목록 (연도별)
pdf_files = {
    # "2021": Path("./dataset/2021_주택임대차분쟁조정사례집.pdf"),
    # "2022": Path("./dataset/2022_주택임대차분쟁조정사례집.pdf"),
    # "2023": Path("./dataset/2023_주택임대차분쟁조정사례집.pdf"),
    # "2025": Path("./dataset/2025_주택임대차분쟁조정사례집.pdf"),
    "law": Path("./dataset/주택임대차보호법.pdf")
}

base_out_dir = Path("pages")
base_out_dir.mkdir(exist_ok=True)

for year, pdf_path in pdf_files.items():
    print(f"[{year}] 처리 시작:", pdf_path)

    # 연도별 출력 디렉토리
    out_dir = base_out_dir / year
    out_dir.mkdir(exist_ok=True)

    doc = fitz.open(pdf_path)

    for i, page in enumerate(doc):
        pix = page.get_pixmap(dpi=300)  # OCR용 권장 DPI
        out_path = out_dir / f"page_{i+1:03}.png"
        pix.save(out_path)

    print(f"[{year}] 총 {len(doc)} 페이지 이미지 추출 완료")

print("모든 PDF 처리 완료")
```
